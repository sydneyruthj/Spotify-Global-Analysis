## 4. Heatmap using matplotlib

```python
# Create a correlation matrix using pandas

sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(10, 10)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Save heatmap
f.savefig("out.png") 
```

## 5. Heatmap using seaborn

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

```python
# Save heatmap
f.savefig("heatmap.png") 
```

#### There are not many strong correlative relationships in this set, but there are a couple of interest. I was hoping to find more relationships between the music descriptions and the popularity metric. Speechiness has the closest linear relationship with popularity (less spoken word is associated with more popular tracks). Other significant relationships include acousticness with energy and loudness. The negative relationship suggest that the more louder and more energetic a song is, the less likely it is that the song is acoustic, which makes sense. We can also see that valence (which is a measure of happiness) is associated with danceability, energy, and loudness, which makes sense. 

## 6. Scatterplots

```python
# Create a scatterplot for the "energy" and "acousticness" columns in seaborn

sns.lmplot(x = 'acousticness', y = 'energy', data = sub)
```

#### There is a clear relationship between acousticness and energy. Around a 0.6 value for acousticness, the energy of a song severely cuts off. Songs that are very acoustic rarely have an energy value over 0.4.

```python
# Create a scatterplot for the "popularity" and "speechiness" columns in seaborn

sns.lmplot(x = 'speechiness', y = 'popularity', data = sub)
```

#### The tracks that are high on speechiness are more likely to be podcasts or audiobooks, so it makes sense that they would not be as popular as music. While high speechiness is a good predictor for low popularity, low speechiness is not a good predictor for high popularity.

## 11. Grouping by name_count

```python
# Create a new column with the count of each item (ChaptGPT)
df['song_count'] = df.groupby('name')['name'].transform('count')

print(df)
```

```python
# Create a scatterplot for the "energy" and "acousticness" columns in seaborn

sns.lmplot(x = 'song_count', y = 'popularity', data = df)
```

```python
subpop = df[['popularity', 'song_count']]

subpop.corr()
```

#### As I suspected, there is a positive correlation between the number of time a song appears in this dataframe and the Spofify popularity metric
